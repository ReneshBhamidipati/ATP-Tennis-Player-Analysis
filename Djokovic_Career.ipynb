```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# In this portion of the analysis, I will focus on Novak's Win/Loss for the first 10 years of his career, season by season.

prefix = '/Users/reneshbhamidipati/Udemy_Python/ATP_Matches/atp_matches_'
suffix = '.csv'

file_names = []
for i in range(2005, 2016):
    file_name = prefix + str(i) + suffix
    file_names.append(pd.read_csv(file_name))

atp_matches = pd.concat(file_names, ignore_index = True)

player_name = 'Novak Djokovic'
df_novak = atp_matches.loc[(atp_matches['winner_name'] == player_name)|(atp_matches['loser_name'] == player_name)].copy()
df_novak_win = atp_matches.loc[(atp_matches['winner_name'] == player_name)].copy()
df_novak_loss = atp_matches.loc[(atp_matches['loser_name'] == player_name)].copy()

df_novak_win.loc[:,'tourney_date'] = pd.to_datetime(df_novak_win['tourney_date'], format='%Y%m%d')
df_novak_loss.loc[:,'tourney_date'] = pd.to_datetime(df_novak_loss['tourney_date'], format='%Y%m%d')

years = []
for i in range(5,16):
    years.append(i+2000)
    
annual_wins = df_novak_win.groupby(df_novak_win.tourney_date.dt.year).count().tourney_id
annual_wins
annual_losses = df_novak_loss.groupby(df_novak_loss.tourney_date.dt.year).count().tourney_id 
annual_losses

plt.plot(years,annual_wins,label='Win')
plt.plot(years,annual_losses,label='Loss')
plt.xticks(years, rotation =45)
plt.legend()
```

```python
# From the prior line graph, it is evident that 2011 and 2015 were Novak's standout seasons, due to his high win count and very few losses.
# To get a clearer picture of how dominant he was in those seasons, I designed two pie graphs just to get an idea of his EXACT W/L.

df_novak_15 = pd.read_csv('/Users/reneshbhamidipati/Udemy_Python/ATP_Matches/atp_matches_2015_copy.csv')
df_novak_15

df_novak_11 = pd.read_csv('/Users/reneshbhamidipati/Udemy_Python/ATP_Matches/atp_matches_2011_copy.csv')
df_novak_11

df_merged = pd.concat([
    pd.read_csv('/Users/reneshbhamidipati/Udemy_Python/ATP_Matches/atp_matches_2011_copy.csv'),
    pd.read_csv('/Users/reneshbhamidipati/Udemy_Python/ATP_Matches/atp_matches_2015_copy.csv'),
],ignore_index = True)
df_merged
```

```python
# Given that Novak barely lost in 2011 and 2015 while competing in all 4 Grand Slam tournaments for each respective season,
# I was curious to see how many Grand Slams he won in those two years in comparison to the other years that he won Grand Slams.
# The results weren't surprising...

novak_slams = df_novak_win.loc[df_novak_win.tourney_level == 'G'].groupby(df_novak_win.tourney_id).count()
novak_slams_won = novak_slams.loc[novak_slams.tourney_id == 7]
plt.title("Novak's Grand Slam Count as of 2015", fontsize = 16)
plt.yticks([1,2,3], fontsize = 10)
grandslams = novak_slams_won.groupby(novak_slams_won.index.map(lambda x: x[0:4])).count()
plt.bar(grandslams.index,grandslams.tourney_id, color= 'Gold')
```

```python
# In the last portion of my analysis, I wanted to take a look at which of Novak's opponents gave him the most trouble over the years.
# To achieve this, I felt that a stacked bar chart would depict how many times Novak won (and lost) against his biggest rivals.

loss = df_novak.loc[df_novak.winner_name != 'Novak Djokovic', ['winner_name']]
loss.columns = ['Name']
loss['Status'] = 'Loss'
loss
```

```python
win = df_novak.loc[df_novak.winner_name == 'Novak Djokovic', ['loser_name']]
win.columns = ['Name']
win['Status'] = 'Win'
win
```

```python
novak_opponents = pd.concat([win,loss])
novak_opponents.groupby('Name').count().sort_values('Status', ascending = False)
no_of_matches = novak_opponents.groupby('Name').count().sort_values('Status', ascending = False)
no_of_losses = novak_opponents.loc[novak_opponents.Status == 'Loss',:].groupby('Name').count().sort_values('Status', ascending = False)
```

```python
fig = plt.figure(figsize = (10,10))
plt.xlabel('Opponents', fontsize = 15)
plt.xticks(rotation = 'vertical')
plt.ylabel('Number of Matches', fontsize = 15)
plt.title("Novak's head-to-head against his opponents", fontsize = 20)
plt.bar(no_of_matches.index[0:10],no_of_matches.Status[0:10],label = 'Win', color = 'Green')
plt.bar(no_of_losses.index[0:10],no_of_losses.Status[0:10],label = 'Loss', color = 'Red')
plt.legend()
```

```python
# After analyzing the stacked bar chart above, we can conclude that Rafael Nadal, Roger Federer and Andy Murray 
# have given Novak the hardest time in his career, given that he has lost to those 3 players more than any of his other opponents.
```
